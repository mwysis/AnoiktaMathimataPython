## 3.1 Ορισμοί

**Μεταβλητή (variable)** είναι το όνομα κάποιου "πράγματος" (δηλαδή, η μεταβλητή είναι κάποιο αναγνωριστικό), το οποίο:
* **κρατείται (αγκαζάρεται) στη μνήμη** (για -σε μονοσήμαντη αντιστοιχία με- το "πράγμα" αυτό)
* και στο οποίο καταχωρείται κάποια ιδιαίτερη **τιμή (value)** (σε αντίστοιχο χώρο και θέση στη μνήμη). 

**Καταχώρηση (assignment)** ονομάζεται η αντιστοίχηση τιμών σε μεταβλητές.

**Τελεστής (operator)** ονομάζεται ο οποιοσδήποτε μετασχηματισμός των τιμών μιας ή περισσότερων μεταβλητών σύφωνα με μαθηματικές πράξεις (αριθμητικές κλπ.) ή λογικούς κανόνες (συνδετικούς, συμπερασματικούς κλπ.).

**Έκφραση (expression)** ονομάζεται ο οποιοσδήποτε συνδυασμός μεταβλητών, τιμών και τελεστών.

**Δήλωση (statement)** είναι κάτι (κάποιο "πράγμα") που μπορεί η Python να εκτελέσει.

Μια σειρά (αλληλουχία) από δηλώσεις λέγεται ότι σχηματίζει ένα **σκριπτ (script)**. Η Python το διερμηνεύει και το εκτελεί με τη σειρά που οι δηλώσεις του σκριπτ είναι γραμμένες.

Η έξοδος (output) της εκτέλεσης μιας δήλωσης (ή ενός σκριπτ ή κάποιου μέρους του σκριπτ)
* είτε προβάλλεται στην οθόνη (π.χ., το αποτέλεσμα κάποιου υπολογισμού)
* ή όχι (π.χ., μια νέα καταχώρηση που κρατείται ή αλλάζει κάτι στη μνήμη)

```python
counter = 100          # Καταχώρηση ακέραιας τιμής σε μεταβλητή
miles   = 1000.0       # Καταχώρηση αριθμού κινητής υποδιαστολής (float) σε μεταβλητή
name    = "John"       # Καταχώρηση συμβολοσειράς σε μεταβλητή

# Τρεις εκφράσεις
17
counter
counter + 17

# Τρεις δηλώσεις
print counter
print miles
print name

a = b = c = 1           # Πολλαπλή καταχώρηση της ίδιες τιμής σε περισσότερες της μιας μεταβλητές
a, b, c = 1, 2, "john"  # Πολλαπλή καταχώρηση πολλαπλών τιμών σε περισσότερες της μιας μεταβλητές
a, b = b, a
```

## 3.2 Τύποι Δεδομένων

**Δεδομένα (data)** λέγονται οι τιμές των μεταβλητών ενός προγράμματος είτε ως είσοδος (input) ή ως έξοδος (output) της εκτέλεσης του προγράμματος.

Συχνά, ταυτίζοντας τις μεταβλητές με τις τιμές που παίρνουν, θα μιλάμε για **τύπους μεταβλητών**.

Οι πέντε **κύριοι τύποι δεδομένων** της Python είναι:
1. **Αριθμοί**
2. **Συμβολοσειρές**
3. Λίστες
4. Πλειάδες
5. Λεξικά

[Για αριθμούς (και αριθμητικές πράξεις) θα μιλήσουμε στην επόμενη ενότητα, ενώ για συμβολοσειρές (και πράξεις σ' αυτές) στην μεθεπόμενη. Οι 3 άλλοι τύποι δεδομένων θα εξετασθούν διαδοχικά στις υπόλοιπες ενότητες που ακολουθούν.]

Οι τέσσερις (κύριοι) **τύποι αριθμών** είναι οι εξής:
1. **int**: ακέραιοι αριθμοί (θετικοί ή αρνητικοί) 
2. **long**: *ακέραιοι* αριθμοί με μεγάλο πλήθος ψηφίων (στο δεκαδικό ή το οκταδικό ή το δεκαεξαδικό σύστημα)
3. **float**: πραγματικοί (μη ακέραιοι) αριθμοί (ή αριθμοί κινητής υποδιαστολής) 
4. **complex**: μιγαδικοί αριθμοί (της μορφής x + yj, όπου x, y είναι παραγματικοί-float)

## 3.3 Αναπαραστάσεις και Μετατροπές μεταξύ Δεδομένων  Διαφόρων Τύπων

### Διάφορες αναπαραστάσεις αριθμών στην Python

```python
u = 10000
u1 = 10**4
u2 = 1e+4
print 'The integer 10**4 is represented (written) either as', u1, 'or as 1e+4'
print 'However, 10**4 and 10000 are of', type(u1), type(u), 'but', '1e+4 is of', type(u2)
print 'The float corresponding to', u,'is represented (written) as', float(u)
print

v = 101.35
v1 = 10135*10**(-2)
v2 = 10135e-2
print 'The float', v, 'is represented (written) either as 10135*10**(-2) or as 10135e-2'
print 'All', v, ', 10135*10**(-2), 10135e-22 are of', type(v), type(v1), type(v2)
print 'The integer corresponding to', v,'is represented (written) as', int(v)
print

x = 10**100
x1 = 10**100L
x2 = 1e+100
print 'The long integer 10**100 is represented (written) either as 10**100L or as 1e+100' 
print 'Note that 10**100 =', x1, 'is', x == x1
print 'However, 10**100 and 10**100L are of', type(x), type(x1), 'but 1e+100 is of', type(x2)
print 'Therefore,', x, '= 1e+100 is', x == x2, 'and 10**100L = 1e+100 is', x1 == x2
# "==" είναι ένας λογικός τελεστής (Boole) [Θα το δούμε στη συνέχεια]
print 'The integer corresponding to 10**100 is', int(x)
print 'The float corresponding to 10**100', 'is', float(x)
print 'However, when printing the integer corresponding to 1e+100, accuracy is lost and we get the number', int(x2)
print 'Note that int(1e+100) = 10**100 is', x == int(x2)
```

### Μετατροπές μεταξύ τύπων αριθμών

```python
k = 10**50
l = long(k)
m = int(k)
n = float(k)
print '10**50 is printed as', k
print '10**50 is of', type(k)
print 'long(10**50) is printed as', l
print 'int(10**50) is printed as', m
print 'float(10**50) is printed as', n
print 'Note that 10**50 = long(10**50) is', k == l 
print 'Note that 10**50 = int(10**50) is', k == m 
print 'Note that long(10**50) = int(10**50) is', l == m 
print 'But that long(10**50) = float(10**50) is', l == n 
print

x = 16.7825
y = long(x)
z = int(x)
print '16.7825 is of', type(x)
print 'long(16.7825) is printed as', y
print 'int(16.7825) is printed as', z
print 'Note that 16.7825 = long(16.7825) is', x == y 
print 'Note that 16.7825 = int(16.7825) is', x == z 
print 'But that long(16.7825) = int(16.7825) is', y == z 
```

##### Πράξεις μεταξύ δεδομένων διαφορετικού τύπου απαγορεύονται και δίνουν λάθος!

```python
a = str(x)
print 'str(16.7825) =', a
print 'str(16.7825) is of', type(a)
print '16.7825 + 10 =', x + 10
print 'str(16.7825) + 10 =', a + 10
```

### Μετατροπές μεταξύ αριθμών και συμβολοσειρών

#### Για συμβολοσειρές συμβόλων/γραμμάτων (όχι ψηφίων/bits)

```python
a = 'Patra2000'                    # Οι συμβολοσειρές πάντα μέσα σε (μονά) '.' 
                                   # ή (διπλά) "," [Θα τα δούμε στη συνέχεια]
print "'Patra200' is of", type(a)
```

##### Συμβολοσειρές που δεν αποτελούνται μόνο από ψηφία/bits ΔΕΝ μετατρέπονται σε αριθμούς!

```python
ia = int(a)
print ia
```

#### Για συμβολοσειρές που αποτελούνται από ψηφία/bits

```python
print 'From numbers to strings:'
x = 1509.07
sx = str(x)
print x, 'is of', type(x)
print "'1509.07' = str(1509.07) is of", type(sx)
print

print 'From strings to numbers:'
s = '1509.07'
fs = float(s)
print "'1509.07' is of", type(s)
print fs, "= float('1509.07') is of", type(fs)
print
print '''"type(1509.07) = float" is''', type(1509.07) == float 
print '''"type('1509.07') = float" is''', type('1509.07') == float 
print '''"type('1509.07') = str" is''', type('1509.07') == str 
```

#### Μετατροπές αριθμών από συμβολοσειρές ως προς οποιαδήποτε βάση σε ακέραιους δεκαδικούς αριθμούς

```python
a = '010101'
print "'010101' is of", type(a)
print "int('010101') is of", type(int(a))
a1 = int(a,2)
print "'010101' is a number in base = 2 and int('010101') as a decimal number is", a1
print "'321' is a number in base = 4 and int('321') as a decimal number is", int('321',4)
print "'2ac' is a number in base = 16 and int('2ac') as a decimal number is", int('2ac',16)
```

```python
# Θα χρησιμοποιήσουμε την βιβλιοθήκη της Python που δίνει ώρα και ημερομηνία:

import datetime
now_today = datetime.datetime.now()
print 'now_today is of', type(now_today)
print '''"type(now_today) = str" is:'''
print type(now_today) == str
snt = str(now_today)
print 'Now today as a string is:'
print snt
rnt = repr(now_today)
print 'Now today as an expression is:'
print rnt
ent = eval(rnt)
print 'Now today as a string obtain from the corresponding expression is:'
print ent
```
