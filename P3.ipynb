## 3.1 Ορισμοί

**Μεταβλητή (variable)** είναι το όνομα κάποιου "πράγματος" (δηλαδή, η μεταβλητή είναι κάποιο αναγνωριστικό), το οποίο:
* **κρατείται (αγκαζάρεται) στη μνήμη** (για -σε μονοσήμαντη αντιστοιχία με- το "πράγμα" αυτό)
* και στο οποίο καταχωρείται κάποια ιδιαίτερη **τιμή (value)** (σε αντίστοιχο χώρο και θέση στη μνήμη). 

**Καταχώρηση (assignment)** ονομάζεται η αντιστοίχηση τιμών σε μεταβλητές.

**Τελεστής (operator)** ονομάζεται ο οποιοσδήποτε μετασχηματισμός των τιμών μιας ή περισσότερων μεταβλητών σύφωνα με μαθηματικές πράξεις (αριθμητικές κλπ.) ή λογικούς κανόνες (συνδετικούς, συμπερασματικούς κλπ.).

**Έκφραση (expression)** ονομάζεται ο οποιοσδήποτε συνδυασμός μεταβλητών, τιμών και τελεστών.

**Δήλωση (statement)** είναι κάτι (κάποιο "πράγμα") που μπορεί η Python να εκτελέσει.

Μια σειρά (αλληλουχία) από δηλώσεις λέγεται ότι σχηματίζει ένα **σκριπτ (script)**. Η Python το διερμηνεύει και το εκτελεί με τη σειρά που οι δηλώσεις του σκριπτ είναι γραμμένες.

Η έξοδος (output) της εκτέλεσης μιας δήλωσης (ή ενός σκριπτ ή κάποιου μέρους του σκριπτ)
* είτε προβάλλεται στην οθόνη (π.χ., το αποτέλεσμα κάποιου υπολογισμού)
* ή όχι (π.χ., μια νέα καταχώρηση που κρατείται ή αλλάζει κάτι στη μνήμη)

counter = 100          # Καταχώρηση ακέραιας τιμής σε μεταβλητή
miles   = 1000.0       # Καταχώρηση αριθμού κινητής υποδιαστολής (float) σε μεταβλητή
name    = "John"       # Καταχώρηση συμβολοσειράς σε μεταβλητή

# Τρεις εκφράσεις
17
counter
counter + 17

# Τρεις δηλώσεις
print counter
print miles
print name

a = b = c = 1           # Πολλαπλή καταχώρηση της ίδιες τιμής σε περισσότερες της μιας μεταβλητές
a, b, c = 1, 2, "john"  # Πολλαπλή καταχώρηση πολλαπλών τιμών σε περισσότερες της μιας μεταβλητές

## 3.2 Τύποι Δεδομένων

**Δεδομένα (data)** λέγονται οι τιμές των μεταβλητών ενός προγράμματος είτε ως είσοδος (input) ή ως έξοδος (output) της εκτέλεσης του προγράμματος.

Οι πέντε **κύριοι τύποι δεδομένων** της Python είναι:
* Αριθμοί
* Συμβολοσειρές
* Λίστες
* Πλειάδες
* Λεξικά

Οι τέσσερις **κύριοι τύποι αριθμών** είναι οι εξής:
* Ακέραιοι αριθμοί (θετικοί ή αρνητικοί) (**int**)
* Αριθμοί με μεγάλο πλήθος ψηφίων (στο δεκαδικό ή το οκταδικό ή το δεκαεξαδικό σύστημα)(**long**)
* Πραγματικοί αριθμοί (ή αριθμοί κινητής υποδιαστολής) (**float**)
* Μιγαδικοί αριθμοί (της μορφής x + yj, όπου x, y είναι παραγματικοί)(**complex**)
